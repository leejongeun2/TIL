```python
import seaborn as sns # 그래프

# 1. 경향성 표현하기
sns.set_theme(style='whitegrid')
penguins = sns.load_dataset("penguins").dropna()
# print(penguins)

# lineplot => 어떤 지표가 커짐에 따라서, 나머지가 어떻게 변화하는지, 시간에 따른 변화량 볼때등등

result = sns.lineplot(data=penguins, x="body_mass_g", y="bill_length_mm", ci=None, hue="species")
print(result)
```

```python
sns.pointplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette="Set2")
```

```python
# 펭귄데이터에 histplot을 출력: 연속적인 변나눠수를 구간으로 서 데이터를 카운트함, y축 고정
sns.histplot(data=penguins, x="bill_depth_mm", bins=30, hue="species", multiple="stack") # 보라색은 겹친 것임
```

```python
sns.barplot(data=penguins, x="island", y="body_mass_g", hue="sex", palette="Set2") # 막대그래프인데, 경쟁사별 매출차이(검은색은 통계적 오차)
```

```python
sns.countplot(data=penguins, x="species") # 카테고리가 몇개씩 있는지 써주는 것
```

```python
sns.boxplot(data=penguins, x="species", y="body_mass_g") # 박스 한가운데 중앙값, 어떤 데이터를 오름차순으로 정렬해서 가운데값, 아래는 25%, 위는 75%
# 다이아몬드는 이상치, 중요!!
# 데이터의 각 종류별로 사분위 수를 표시하는 plot
# 특정 데이터의 전체적인 분포를 확인하기 좋은 시각화 기법
# 박스와 전체 범위의 그림을 통해 outlier를 찾기 쉽습니다.(iqr)
```

```python
sns.swarmplot(data=penguins, x="sex", y="flipper_length_mm")
# 데이터에 대한 분포 자체를 보여주는 plot
# boxplot과 비슷하지만, 전체 분포에 대한 그림을 보여준다는 점에서 다름
# 보통 boxplot과 함께 표시하면, 평균 근처에 데이터가 얼마나 있는지 전체적으로 어떻게 퍼져있는지 모두 확인 가능합니다. 
```

```python
# scatterplot
# lineplot과 비슷하게 x,r에 대한 전체적인 분포를 확인하는 plot
# lineplot은 경향성에 초점을 둔다면, 이건 데이터 그 자체가 퍼져있는 모양에 중점을 둠
# 양의 상간관계, 연관성
sns.scatterplot(data=penguins, x="bill_length_mm", y="flipper_length_mm", hue="species")
```

```python
# 연관성 표현하기
# heatmap, 상간관계
# 정사각형 그림에 데이터에 대한 정도 차이를 색 차이로 보여주느 plot
# 말 그대로 heatmap이기 때문에, 열화상카메라로 사물을 찍은 것처럼 정보의 차이를 보여줌
# pairtplot과 비슷하게 feature간의 관계를 시각화할때 많이 사용한다. 
penguins.corr()
```

```python
sns.heatmap(penguins.corr(), annot=True, cmap="Blues")
```

```python
sns.pairplot(data=penguins, hue="species") # 4개 수치값있으면 쓰기에 괜찮음(6개정도까지 추천)
```
