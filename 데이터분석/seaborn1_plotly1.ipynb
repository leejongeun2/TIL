```python
import seaborn as sns # 

# 1.  
sns.set_theme(style='whitegrid')
penguins = sns.load_dataset("penguins").dropna()
# print(penguins)

# lineplot =>    ,   ,    

result = sns.lineplot(data=penguins, x="body_mass_g", y="bill_length_mm", ci=None, hue="species")
print(result)
```

```python
sns.pointplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette="Set2")
```

```python
#  histplot :      , y 
sns.histplot(data=penguins, x="bill_depth_mm", bins=30, hue="species", multiple="stack") #   
```

```python
sns.barplot(data=penguins, x="island", y="body_mass_g", hue="sex", palette="Set2") # ,  (  )
```

```python
sns.countplot(data=penguins, x="species") #     
```

```python
sns.boxplot(data=penguins, x="species", y="body_mass_g") #   ,     ,  25%,  75%
#  , !!
#       plot
#        
#      outlier  .(iqr)
```

```python
sns.swarmplot(data=penguins, x="sex", y="flipper_length_mm")
#      plot
# boxplot ,       
#  boxplot  ,           . 
```

```python
# scatterplot
# lineplot  x,r     plot
# lineplot   ,        
#  , 
sns.scatterplot(data=penguins, x="bill_length_mm", y="flipper_length_mm", hue="species")
```

```python
#  
# heatmap, 
#          plot
#   heatmap ,       
# pairtplot  feature    . 
penguins.corr()
```

```python
sns.heatmap(penguins.corr(), annot=True, cmap="Blues")
```

```python
sns.pairplot(data=penguins, hue="species") # 4   (6 )
```
